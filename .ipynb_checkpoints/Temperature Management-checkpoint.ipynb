# Temperature Management

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
```

```python
df = pd.read_excel('data\\Room_326.xlsx', index_col=None)
df.set_index('Time');
```

# Plot data

From (Month 7 July) to (Month 10 October)

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize=(18, 5))


axes[0].set_title('Original mean')
axes[0].plot(df['temp. measured'].rolling(250).mean(), label = 'temp measured')
axes[0].plot(df['target temp.'].rolling(250).mean(), label = 'target temp')
axes[0].legend(loc = 3)

df['Ticks'] = range(0, len(df.index.values))
df_one_tenth = df.sample(frac=0.1, random_state=np.random.randint(10))
df_one_tenth = df_one_tenth.sort_values(by=['Ticks'])

axes[1].set_title('One tenth mean')
axes[1].plot(df_one_tenth['temp. measured'].rolling(50).mean())


#axes[2].set_title('One tenth mean')
#axes[2].plot(df_one_tenth['temp. measured'].rolling(4).mean())
```

<H4> Temperature plot </H4>

```python
def plot(temp_measured, target_temp, temp_air_ch1, temp_air_ch2, temp_outside, water_flow_temp_ch1, water_flow_temp_ch2):
    if(temp_measured):
        plt.plot(df['temp. measured'].rolling(250).mean())
    if(target_temp):
        plt.plot(df['target temp.'].rolling(250).mean())
    if(temp_air_ch1):
        plt.plot(df['temp. air-ch.1'].rolling(250).mean())
    if(temp_air_ch2):
        plt.plot(df['temp. air-ch.2'].rolling(250).mean())
    if(temp_outside):
        plt.plot(df['temp. outside'].rolling(250).mean())
    if(water_flow_temp_ch1):
        plt.plot(df['temp. water flow ch.1'].rolling(250).mean())
    if(water_flow_temp_ch2):
        plt.plot(df['temp. water flow ch.2'].rolling(250).mean())
    plt.show()
    
interact(plot, temp_measured=True, target_temp=False, temp_air_ch1=False, temp_air_ch2=False, temp_outside=False, water_flow_temp_ch1=False, water_flow_temp_ch2=False)
```

<H4> Control signals plot </H4>

```python
def plot(heat_control_signal, cooling_control_signal):
    if(heat_control_signal):
        plt.plot(df['heat control signal'].rolling(20).mean(), color='red')
    if(cooling_control_signal):
        plt.plot(df['cooling control signal'].rolling(20).mean(), color='blue')
    plt.show()
    
interact(plot, heat_control_signal=False, cooling_control_signal=False)
```

<H4> Air pressure plot</H4>

```python
def plot(pressure_ch1, pressure_ch2):
    if(pressure_ch1):
        plt.plot(df['pressure air-ch.1'].rolling(80).mean())
    if(pressure_ch2):
        plt.plot(df['pressure air-ch.2'].rolling(80).mean())
    plt.show()
interact(plot, pressure_ch1=False, pressure_ch2=False)
```
