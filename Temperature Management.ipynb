# Temperature Management

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('data\\Room_326.xlsx', index_col=None)
df.set_index('Time');
```

# Plot data

From (Month 7 July) to (Month 10 October)

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize=(18, 5))


axes[0].set_title('Original mean')
axes[0].plot(df['temp. measured'].rolling(250).mean(), label = 'temp measured')
axes[0].plot(df['target temp.'].rolling(250).mean(), label = 'target temp')
axes[0].legend(loc = 3)

df['Ticks'] = range(0, len(df.index.values))
df_one_tenth = df.sample(frac=0.1, random_state=np.random.randint(10))
df_one_tenth = df_one_tenth.sort_values(by=['Ticks'])

axes[1].set_title('One tenth mean')
axes[1].plot(df_one_tenth['temp. measured'].rolling(50).mean())


#axes[2].set_title('One tenth mean')
#axes[2].plot(df_one_tenth['temp. measured'].rolling(4).mean())
```
