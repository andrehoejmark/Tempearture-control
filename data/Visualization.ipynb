# Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('data\\Room_326.xlsx', index_col=None)
df.set_index('Time');
df.head()
```

```python
df = df.iloc[10000:15000]

fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize=(14, 8))
df['temp. measured'].plot(ax=axes[0,0], title ='temp. measured')
df['target temp.'].plot(ax=axes[0,1], title ='target temp')
df['heat control signal'].plot(ax=axes[1,0], title ='heat control signal')
df['cooling control signal'].plot(ax=axes[1,1], title ='cooling control signal')


```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize=(14, 8))
df['temp. air-ch.1'].plot(ax=axes[0,0], title ='temp. air-ch.1')
df['pressure air-ch.1'].plot(ax=axes[0,1], title ='pressure air-ch.1')
df['temp. air-ch.2'].plot(ax=axes[1,0], title ='temp. air-ch.2')
df['pressure air-ch.2'].plot(ax=axes[1,1], title ='pressure air-ch.2')
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize=(14, 8))
df['temp. water flow ch.1'].plot(ax=axes[0,0], title ='temp. water flow ch.1')
df['temp. water flow ch.2'].plot(ax=axes[0,1], title ='temp. water flow ch.2')
df['temp. outside'].plot(ax=axes[1,0], title ='temp. outside')
```
